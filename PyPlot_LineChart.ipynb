```python
import matplotlib.pyplot as plt
```

```python
savings = [0,100,150,200,300,500,300,350,500,700,750,1000]
months = ['Jan', 'Feb','Mar','Apr','May','June','July','Aug','Sept','Oct','Nov','Dec']
```

```python
plt.plot(months,savings)
plt.xlabel('Month')
plt.ylabel('Savings')
plt.title('Savings Results')
plt.show()
```
